# 3.데이터 준비

- **전처리** 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

<img src = "https://github.com/Jangrae/img/blob/master/code_13.png?raw=true" width=700 align="left"/>

**1) 결측치 처리**

- 결측치가 있으면 제거하거나 적절한 값으로 채웁니다.

```python
data_bu = data.copy()
```

```python
# 결측치 확인
data = data_bu.copy()
data.isna().sum()
```

```python
# 전날 값으로 결측치 채우기
data.fillna(method = 'ffill', inplace= True)

# 확인
data.isna().sum()
```

**2) 변수 제거**

- 분석에 의미가 없다고 판단되는 변수는 제거합니다.

```python
# 변수 제거
data.drop(['Month', 'Day'], axis = 1, inplace= True)


# 확인
data.head()
```

**3) x, y 분리**

- target 변수를 명확히 지정합니다.
- target을 제외한 변수를 x 데이터프레임으로 선언합니다.
- y 데이터프레임은 target 변수만을 가집니다.

```python
# target 확인
target = 'Ozone'

# 데이터 분리
x = data.drop('Ozone', axis= 1)
y = data.loc[:, target]
```

```python
# display(x)
# display(y)
```

**4) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

<img src = "https://github.com/Jangrae/img/blob/master/split.png?raw=true" width=500 align="left"/>

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

# 4.모델링

- 본격적으로 모델을 **선언**하고 **학습**하고 **평가**하는 과정을 진행합니다.
- 우선 **회귀** 문제인지 **분류** 문제인지 명확히 구분합니다.

<img src = "https://github.com/Jangrae/img/blob/master/code_14.png?raw=true" width=700 align="left"/>

**1) 모델링**

- 회귀 문제 인가요? 분류 문제인가요?
- 회귀인지 분류인지에 따라 사용할 알고리즘과 평가 방법이 달라집니다.
- 우선 다음 알고리즘과 평가 방법을 사용합니다.
    - 알고리즘: LinearRegression
    - 평가방법: mean_absolute_error

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
# mean_absolute_error : 평가방법 : 평균의 절대적 오차
```

```python
# 2단계: 선언하기
# 선형 회귀 모델
model = LinearRegression()
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(x_test)
```

```python
# 5단계: 평가하기
print(mean_absolute_error(y_test, y_pred))
```

```python
# 결과 시각화

plt.figure(figsize = (6, 3))

plt.plot(y_pred, label = 'y_pred')
plt.plot(y_test.values, label = 'y_test')
plt.axhline(y_test.mean(), color = 'r', linestyle = '--')
plt.legend()
plt.grid()
plt.show()
```
